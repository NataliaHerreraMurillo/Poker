```python
#https://www.youtube.com/watch?v=rONhdonaWUo
import bs4
import requests
from bs4 import BeautifulSoup

def parsePrice():
    r1 = requests.get("https://finance.yahoo.com/quote/NVDA?p=NVDA")
    soup1 = bs4.BeautifulSoup(r1.text, "lxml")
    price1 = soup1.find_all("div", {'class': 'My(6px) Pos(r) smartphone_Mt(6px)'})[0].find('span').text
    
    r2 = requests.get("https://finance.yahoo.com/quote/FB/")
    soup2 = bs4.BeautifulSoup(r2.text, "lxml")
    price2 = soup2.find_all("div", {'class': 'My(6px) Pos(r) smartphone_Mt(6px)'})[0].find('span').text
    
    valores_nuevos =float (price1)
    valores_nuevos2 =float(price2)
    
    return (valores_nuevos, valores_nuevos2)
```

```python
while True:
    print(parsePrice())
```
